```python
d={'Delhi Daredevils':'Delhi Capitals',
  'Kings XI Punjab':'Punjab Kings',
  'Rising Pune Supergiant': 'Pune Warriors',
   'Rising Pune Supergiants': 'Pune Warriors',
  'Gujarat Lions':'Gujarat Titans'}
df= df.replace(d.keys(),d.values())
df.nunique()
```

```python
teams = df.Team1.unique()
l=[]
for team in teams:
    played =df[(df.Team1 == team)|(df.Team2==team)].shape[0]
    win = df[df.WinningTeam ==team].shape[0]
    homeWin = df[(df.WinningTeam ==team) & (df.Team1==team)].shape[0]
    homePlay = df[df.Team1==team].shape[0]
    awayWin = df[(df.WinningTeam ==team) &(df.Team2 ==team)].shape[0]
    awayPlay = df[df.Team2 == team].shape[0]
                  
    l.append([team,played,(win/played)*100, (homeWin/homePlay)*100,(awayWin/awayPlay)*100])
    
new_df = pd.DataFrame(l,columns=['Team','play','Win','Home-Win','Away-Win'])
```

```python
df = pd.read_csv("diabetes_unclean.csv")
df.info()
```

```python
df.describe(exclude=np.number)
```

```python
#find the number of missing data of each column : 
df.isna().sum()
```

```python
df['HDL'] = df['HDL'].fillna(df['HDL'].median())
print(df['HDL'])

```

```python
df['AGE'] = df['AGE'].fillna(df['AGE'].mean())
print(df['AGE'])
```

```python
df = df.dropna()
```

```python
import matplotlib.pyplot as plt 
plt.boxplot(df['Cr'])
plt.show()
```

```python
plt.boxplot(df['Cr'])
plt.show()
```

```python
q1 = df['Cr'].quantile(0.25)
q3 = df['Cr'].quantile(0.75)
iqr = q3-q1
ul = q3 + 1.5 * iqr
ll = q1 - 1.5 * iqr
df =df[(df['Cr']>=ll)&(df['Cr']<=ul)]
print(df)
```
